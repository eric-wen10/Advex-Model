```python
# Gets images and .xml files with knives in them
import os
import shutil
import xml.etree.ElementTree as ET

# Set the paths to the input and output folders
input_folder = "/Users/ericwen10/Advex/Knife-Model/USRT/USRT Images/"
output_folder = "/Users/ericwen10/Advex/Knife-Model/USRT/USRT Images Cleaned Knife/"

# Loop through each file in the input folder
for filename in os.listdir(input_folder):
    if filename.endswith(".xml"):
        # Get the corresponding .jpg filename
        jpg_filename = filename.replace(".xml", ".jpg")
        # Check if the .xml file contains an object with name "Knife"
        with open(os.path.join(input_folder, filename), "r") as f:
            xml_tree = ET.parse(f)
            root = xml_tree.getroot()
            contains_knife = any(obj.find("name").text == "Person" for obj in root.findall("object"))
            if contains_knife:
                # If the .xml file contains an object with name "Knife", remove any other objects from the file
                for obj in root.findall("object"):
                    if obj.find("name").text != "Knife":
                        root.remove(obj)
                # Write the modified XML file to the output folder
                xml_tree.write(os.path.join(output_folder, filename))
                # Copy the corresponding .jpg file to the output folder
                shutil.copy(os.path.join(input_folder, jpg_filename), os.path.join(output_folder, jpg_filename))
                print(f"Moved {filename} and {jpg_filename} to {output_folder}")
```

```python
#Given a folder with xml files, split into three folders for train, val, test (75,15,5)
import random

# Set the path to the folder containing the .xml files
xml_folder = "/Users/ericwen10/Advex/Knife-Model/USRT/Cleaned Knife XML/"

# Set the path to the destination folders for train, val, and test
train_folder = "/Users/ericwen10/Advex/Knife-Model/USRT/Train/"
val_folder = "/Users/ericwen10/Advex/Knife-Model/USRT/Val/"
test_folder = "/Users/ericwen10/Advex/Knife-Model/USRT/Test/"

# Set the ratios for train, val, and test
train_ratio = 0.75
val_ratio = 0.2
test_ratio = 0.05

# Create the destination folders if they don't already exist
os.makedirs(train_folder, exist_ok=True)
os.makedirs(val_folder, exist_ok=True)
os.makedirs(test_folder, exist_ok=True)

# Get a list of all the .xml files in the xml_folder
xml_files = [f for f in os.listdir(xml_folder) if f.endswith('.xml')]

# Shuffle the xml_files list
random.shuffle(xml_files)

# Get the number of xml_files
num_files = len(xml_files)

# Calculate the number of files for each split
num_train = int(num_files * train_ratio)
num_val = int(num_files * val_ratio)
num_test = int(num_files * test_ratio)

# Split the files into train, val, and test
train_files = xml_files[:num_train]
val_files = xml_files[num_train:num_train+num_val]
test_files = xml_files[num_train+num_val:]

# Move the files into the respective folders
for file in train_files:
    shutil.move(os.path.join(xml_folder, file), os.path.join(train_folder, file))

for file in val_files:
    shutil.move(os.path.join(xml_folder, file), os.path.join(val_folder, file))

for file in test_files:
    shutil.move(os.path.join(xml_folder, file), os.path.join(test_folder, file))


```

```python
#Given an xml file, I'm creating annotation.json files for training, validation, and test
import os
import json
from xml.etree import ElementTree as ET

folder_path = "/Users/ericwen10/Advex/Knife-Model/USRT/Test/"  # specify folder path here
categories = [{"supercategory": "none", "id": 2, "name": "knife"}]
images = []
annotations = []

image_id = 1
annotation_id = 1

for filename in os.listdir(folder_path):
    if filename.endswith(".xml"):
        xml_file = os.path.join(folder_path, filename)
        tree = ET.parse(xml_file)
        root = tree.getroot()
        image_filename = root.find("filename").text
        image_path = root.find("path").text
        image_width = int(root.find("size/width").text)
        image_height = int(root.find("size/height").text)
        image_dict = {"file_name": image_filename, "height": image_height, "width": image_width, "id": image_filename}
        images.append(image_dict)
        for obj in root.findall("object"):
            category = obj.find("name").text
            bbox = obj.find("bndbox")
            bbox_list = [float(bbox.find("xmin").text), float(bbox.find("ymin").text),
                         float(bbox.find("xmax").text) - float(bbox.find("xmin").text),
                         float(bbox.find("ymax").text) - float(bbox.find("ymin").text)]
            annotation_dict = {"area": bbox_list[2] * bbox_list[3], "iscrowd": 0, "bbox": bbox_list,
                               "category_id": categories[0]["id"],
                               "ignore": 0, "segmentation": [], "image_id": image_filename, "id": annotation_id}
            annotations.append(annotation_dict)
            annotation_id += 1
        image_id += 1

data_dict = {"type": "instances", "categories": categories, "images": images, "annotations": annotations}

json_path = "/Users/ericwen10/Advex/Knife-Model/USRT/Test/Test annotations.json"  # specify output JSON file path here
with open(json_path, "w") as f:
    json.dump(data_dict, f, indent=4)
```
